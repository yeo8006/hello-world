```python
def compute(n):
	ans = 0
	x = 1  # Represents the current Fibonacci number being processed
	y = 2  # Represents the next Fibonacci number in the sequence
	while x <= n:
		if x % 2 == 0:
			ans += x
		x, y = y, x + y
	return str(ans)
```

```python
compute(4000000)
```
